```python
series_log = data['log_passengers']
train = series_log[:-24]
test = series_log[-24:]
```

```python
from statsmodels.tsa.arima.model import ARIMA
arima_model = ARIMA(train, order = (1,1,1))
arima_fit = arima_model.fit()
```

```python
arima_pred_log = arima_fit.forecast(steps=len(test))

arima_pred = np.exp(arima_pred_log)
test_actual = np.exp(test)

plt.figure(figsize=(12, 5))
plt.plot(test_actual.index, test_actual, label='Actual passengers')
plt.plot(test_actual.index, arima_pred, label='ARIMA predicted passengers')
plt.title('ARIMA: Actual vs Predicted')
plt.xlabel('Month')
plt.ylabel('#Passengers')
plt.legend(loc='best')
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
sarima_model = SARIMAX(
    train,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_fit = sarima_model.fit(disp=False)
```

```python
sarima_pred_log = sarima_fit.forecast(steps=len(test))

sarima_pred = np.exp(sarima_pred_log)

plt.figure(figsize=(12, 5))
plt.plot(test_actual.index, test_actual, label='Actual passengers')
plt.plot(test_actual.index, arima_pred, label='ARIMA predicted')
plt.plot(test_actual.index, sarima_pred, label='SARIMA predicted')
plt.title('ARIMA vs SARIMA: Actual vs Predicted')
plt.xlabel('Month')
plt.ylabel('#Passengers')
plt.legend(loc='best')
plt.show()
```
